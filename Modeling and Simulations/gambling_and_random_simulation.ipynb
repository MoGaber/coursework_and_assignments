```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def biased_dice():
    generate = random.random()
    if generate <= 0.6:
        return "H"
    else:
        return "T"
```

We can assume that the walker starts at 0, Heads will make him go 1 and tails will make him go -1

```python
random_walks = []
for i in range(200):
    random_walks.append([0]) # will make a list for each random walk
    for j in range(30):
        roll_dice = biased_dice()
        if roll_dice == 'H':
            random_walks[-1].append(random_walks[-1][-1]+1)
        else:
            random_walks[-1].append(random_walks[-1][-1]-1)
```

```python
plt.plot([i for i in range(31)],random_walks[0])
plt.xlabel('Time')
plt.ylabel("path")
```

```python
means = [np.mean(i) for i in random_walks]
```

```python
plt.hist(means)
plt.title("Histogram of 200 random walks means")
```

```python
print("Varinace: ", np.std(means))
```

#### The Gambler Game

```python
def gambling(bound):
    gambler = 100
    house = 2000
    gambler_money = [gambler]
    for i in range(bound):
        bet  =random.random()
        if bet <=0.5:
            gambler+=1
            house -=1
        else:
            gambler-=1
            house+=1
        gambler_money.append(gambler)
    
    
    plt.plot(range(bound+1), gambler_money)
    plt.xlabel('Time steps')
    plt.ylabel("Gambler's money")
    plt.title(f"Gambler Money over {bound} steps")
    plt.show()
    return(gambler_money, bound)
for bound in [400, 600, 800, 1000]:
    gambling(bound)
```
