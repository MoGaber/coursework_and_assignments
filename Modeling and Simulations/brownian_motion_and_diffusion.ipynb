```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
import math
```

```python
#ex4 - Part a
walks = []
iterations = 10000
for j in range(100):
    current = 0
    all_steps =[]
    for i in range(iterations):
        if current == -4:
            current = -3
        step = random.choice([-1, 1])
        current+= step
        all_steps.append(current)
    walks.append(all_steps)

```

```python
plt.hist(np.array(walks)[:, -1])
```

```python
plt.plot(walks[0])
```

```python
#ex4 - Part B
walks = []
finals = []
iterations = 10000
for j in range(100):
    current = 0
    all_steps =[]
    for i in range(iterations):
        if current == -4:
            current = -3
        elif current == 6:
            if random.uniform(0, 1) < 0.75:
                current = 5
            else:
                current = 7
        else:
            step = random.choice([-1, 1])
            current+= step
        all_steps.append(current)
    walks.append(all_steps)
    finals.append(all_steps[-1])

plt.hist(np.array(finals))
```

```python
#ex4
current = [0, 0]
steps = 1600

all_walks = []
for i in range(100):
    walks = []
    for i in range(steps):
        degree = np.deg2rad(random.uniform(0, 360))
        step = sts.norm(loc = 0, scale=.5).rvs(1)
        x = step*math.cos(degree)
        y = step*math.sin(degree)
        new_x = current[0]+x
        new_y = current[1]+y
        walks.append([new_x, new_y])
    all_walks.append(walks)
```

```python
plt.scatter(np.array(all_walks)[:, -1][:, 0], np.array(all_walks)[:, -1][:, 1])
```

```python
# Ex9
import numpy as np
import matplotlib.pyplot as plt

steps = 100
trials = 10000  

results = np.zeros((10000, 2), dtype=int)
boundary = 0
hole = 0
for trial in range(trials):
    move = np.zeros((steps + 1, 2), dtype=int)
    for step in range(steps):
        move[step+1] = (
            move[step] +
            [(-1, 0), (1, 0), (0, -1), (0, 1)][np.random.randint(4)]) #directions
        if (move[step+1][0] in [-20, 20]) or (move[step+1][1] in [-20, 20]):
           
           # leaving the boundary
            boundary += 1
            break
        elif (13 <= move[step+1][0] <= 14) and (7 <= move[step+1][1] <= 8):
            hole += 1 #entering the given hole
            break
    results[trial] = move[step+1] 
```

```python
plt.scatter(results[:,0], results[:,1] )
```
